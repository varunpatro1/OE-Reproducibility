```python
rows = data.shape[0]
cols = data.shape[1]
bands = data.shape[2]
```

```python
data_1 = data.reshape((rows*cols, bands))
```

```python
Xsub = data_1[np.arange(0, data_1.shape[0], 40), :]
mu = Xsub.mean(axis = 0)
Cov = np.cov(Xsub, rowvar = False)
ev, evec = eig(Cov)
```

```python
# eigenvector analysis
ev, evec = eig(Cov)
noise = np.zeros(50)
plt.plot(np.arange(50), np.log10(ev[:50]), marker = 'o')
plt.title('Eigenvalue Decay')
plt.xlabel('Component')
plt.ylabel('Log eigenvalue')
plt.legend()
```

```python
fig, axs = plt.subplots(nrows = 3, ncols = 5, figsize = (12,10), constrained_layout = True)
gs = fig.add_gridspec()
X_zeromean = data_1 - mu
X_proj = np.zeros((rows, cols, bands))
r = 0
c = 0
for i in range(0, 15):
    if i == 5:
        r = 1
        c = 0
    if i == 10:
        r = 2
        c = 0
    X_fixed = np.matmul(X_zeromean, evec[:,i])
    X_proj[:, :, i] = X_fixed.reshape((rows, cols))
    axs[r, c].imshow(np.squeeze(X_proj[:, :, i]))
    axs[r,c].set_title('Eigenvalue ' + str(i))
    
    c += 1
```
