```python
subsample_1_ind = np.arange(0,cluster_1.shape[0],4000)
subsample_1 = cluster_1[subsample_1_ind,:].copy()
subsample_2_ind = np.arange(0,cluster_2.shape[0],4000)
subsample_2 = cluster_2[subsample_2_ind,:].copy()
subsample_3_ind = np.arange(0,cluster_3.shape[0],4000)
subsample_3 = cluster_3[subsample_3_ind,:].copy()
subsample_4_ind = np.arange(0,cluster_4.shape[0],4000)
subsample_4 = cluster_4[subsample_4_ind,:].copy()
subsample_5_ind = np.arange(0,cluster_5.shape[0],4000)
subsample_5 = cluster_5[subsample_5_ind,:].copy()
subsample_6_ind = np.arange(0,cluster_6.shape[0],4000)
subsample_6 = cluster_6[subsample_6_ind,:].copy()
subsample_7_ind = np.arange(0,cluster_7.shape[0],4000)
subsample_7 = cluster_7[subsample_7_ind,:].copy()
subsample_8_ind = np.arange(0,cluster_8.shape[0],4000)
subsample_8 = cluster_8[subsample_8_ind,:].copy()
```

```python
print(subsample_1.shape)
print(subsample_2.shape)
print(subsample_3.shape)
print(subsample_4.shape)
print(subsample_5.shape)
print(subsample_6.shape)
print(subsample_7.shape)
print(subsample_8.shape)
```
